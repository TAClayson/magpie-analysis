# Fringe Tracing Algorithum

Add the input file path and run to load the image.

```python
inputFile = "input.JPG"

%matplotlib notebook
import os
import matplotlib.pyplot as plt

if os.path.isfile(inputFile):
    inImage = plt.imread(inputFile)
    image = inImage[:,:,0]/3+inImage[:,:,1]/3+inImage[:,:,2]/3
    fig, ax = plt.subplots()
    plot = ax.imshow(image, cmap = plt.get_cmap('gray'))
    
else:
    print("File not found")
```

Run the code below to perform an FFT of the image.

```python
import numpy as np

fftImage = np.fft.fft2(image)
fftImage = np.fft.fftshift(fftImage)
fftImage = 20*np.log(np.abs(fftImage))

fig, ax = plt.subplots()
plot = ax.imshow(fftImage, cmap = plt.get_cmap('gray'))
```
