```python
%matplotlib notebook
from bdots import ScopeChannel
import matplotlib.pyplot as plt
```

```python
scope="11"
channel="C1"#chico
```

```python
shots=['s1026_15', 's0121_16', 's0429_16']
shots=['s0429_16', 's0505_16', 's0511_16']

sc=[ScopeChannel(s, scope, channel) for s in shots]
```

```python
fig, ax=plt.subplots(figsize=(12,8))
for s in sc:
    ax.plot(s.time, s.data, label=s.shot)
    
ax.set_xlim(0,3000)
ax.legend()
    
```
