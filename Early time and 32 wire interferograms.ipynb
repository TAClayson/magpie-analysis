## s0119_16: 16 wire, gap vs gap. t=168 ns (Chico misfired)

```python
cd "C:\Users\jdavies\Google Drive\MAGPIE\data\2016\s0119_16 C RECONNECTION 12 FRAME"
```

```python
s0119_16_532=NeLMap2('s0119_16 532nm neL.txt', scale=173, rot_angle=-90-20)
s=s0119_16_532
title='s0119_16, t=171ns'
s.data=s.data-3e17
```

```python
s.set_origin(origin=[2906,2182], extent=[-11.5,11.5,-8,8])

fig, ax=plt.subplots(figsize=(8,8))
im=s.plot_data_mm(clim=[0,10], multiply_by=1/1.6e17, ax=ax)
ax.set_title(title, fontsize=28)

ax.set_xlabel('x (mm)',fontsize=28)
ax.set_ylabel('y (mm)',fontsize=28)
ax.set_xticks([-5,0,5])

ax.tick_params(labelsize=28, pad=5, length=10, width=2)


cbar_ax = fig.add_axes([0.80, 0.10, 0.05, 0.8])#l,b,w,h
cbar_ax.tick_params(labelsize=28, pad=5, length=10, width=2)
fig.colorbar(im, cax=cbar_ax, ticks=[0,2,4,6,8,10])
cbar_ax.set_ylabel(r'$n_e$ ($\times$ 10$^{17}$ cm$^{-3}$)',fontsize=28)
```
