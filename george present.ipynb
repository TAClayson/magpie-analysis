```python
%matplotlib inline
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import scipy.special
from scipy import misc
from scipy import ndimage
```

```python
image=misc.imread('coffee_maker.jpg')
image=np.sum(image,2)
```

```python
blurred=ndimage.gaussian_filter(image, sigma=7)
plt.imshow(blurred)
```

```python
background=np.array([[i for j in np.arange(blurred.shape[1])] for i in np.arange(blurred.shape[0])])
```

```python
fig,ax=plt.subplots(figsize=(15,15))
im=ax.imshow(np.cos(background*3e-1+blurred*3e-2), cmap='Greys')
```

```python
im = Image.fromarray(np.uint8(255*(np.cos(background*3e-1+blurred*3e-2))**2))
im.save('george_card.png')
```
