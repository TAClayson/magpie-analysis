```python
%matplotlib notebook
from bdots import ScopeChannel, MitlBdots
import matplotlib.pyplot as plt
import numpy as np

marx_volts={'G':'C1', 'H':'C2','C':'D1', 'Z':'D2'}
mitl_bdots=['A1','A2','B1','B2']
```

```python
scope="11"

marx='Z'
#shots=['s0504_16','s0505_16', 's0607_16']
shots=['s0504_16','s0607_16','s0608_16', 's0609_16']

#bd=[ScopeChannel('s1029_15', scope, c) for c in mitl_bdots]
sc=[ScopeChannel(s, scope, marx_volts[marx]) for s in shots]
#sc[4].time=sc[0].time
```

```python
fig, ax=plt.subplots(figsize=(12,8))
for s in sc:
    ax.plot(s.time, s.data, label=s.shot)
    
ax.set_xlim(0,2000)
#ax.set_ylim(-6,6)
ax.legend()
    
```

```python
mb=[]
#shots=['s0121_16','s0119_16','s0607_16']
shots=['s0607_16','s0119_16','s0120_16', 's0120B16','s0121_16']

for s in shots:
    m=MitlBdots(s)
    m.truncate()
    m.integrate()
    mb.append(m)
```

```python
fig, ax=plt.subplots(figsize=(12,8))
for m in mb:
    b=m.mbds[3]
    ax.plot(b.time_B0, b.B, label=m.shot)
    
ax.set_xlim(0,500)
#ax.set_ylim(-6,6)
ax.legend()
```

```python
print('Shot\t\tBd1\tBd2\tBd3\tBd4\tBTot')
for m in mb:
    stri=m.shot+':\t'
    tot=0
    for b in m.mbds:
        BM=int(np.abs(b.B).max())
        tot=tot+BM
        stri=stri+str(BM)+'\t'
    stri=stri+str(tot)
    print(stri)
```
