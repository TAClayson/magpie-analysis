```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
```

```python


B_0=3
M_A=0.7
L=7e-3
delta=0.7e-3
d_i=0.6e-3


x=np.linspace(-5,5,1000)*L
y=np.linspace(-5,5,1000)*delta
XX, YY=np.meshgrid(x,y)

B=-0.5*M_A*B_0*L/d_i*np.log(np.abs((YY+XX)/(YY-XX)))
```

```python
fig, ax=plt.subplots(figsize=(12,8))
ax.imshow(B, interpolation='none', cmap='bwr', clim=[-50,50])
```
