```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
% cd '../2015/s0513_15 RECONNECTION - Carbon repeat'
data=np.genfromtxt(open('s0513_15 faraday_angle.txt',"r"),delimiter=",")
```

```python
plt.imshow(data[800:1700, 1200:2200], cmap='Greys')
```

```python
avg_rot=np.mean(data[1400:1800, 1200:2200], axis=0)
rot_scale=61 #px/mm
rot_centre=452
```

```python
plt.plot(avg_rot[440:470])
```

```python
ne=np.genfromtxt(open('s0513_15 ne.txt',"r"),delimiter=",")
```

```python
avg_ne=np.mean(ne[200:1500,:], axis=0)
ne_centre=611
ne_scale=100#px per mm
```

```python
rescaling_factor=float(rot_scale)/ne_scale
new_px=np.linspace(0, avg_ne.size,avg_ne.size*rescaling_factor)
old_px=np.arange(avg_ne.size)
avg_ne_scaled=np.interp(new_px, old_px, avg_ne)
ne_clip=avg_ne_scaled.size/2
shift=5
avg_rot_clipped=avg_rot[rot_centre-ne_clip+shift:rot_centre+ne_clip+1+shift]
mag=avg_rot_clipped/avg_ne_scaled
constant=2.62e-13
lamb=1053e-9
alpha=avg_rot_clipped*np.pi/180.0 #convert to radians
B=alpha/(constant*lamb**2*avg_ne_scaled*1e4)
centre_pixel=avg_rot_clipped.size/2+12
max_dist=centre_pixel/float(rot_scale)
mm_scale=np.linspace(-max_dist, max_dist, avg_rot_clipped.size)
```

```python
plt.plot(avg_rot_clipped[360:380])
#plt.plot(avg_ne_scaled*1e-18)
```

```python
fig, ax1=plt.subplots(figsize=(15,15))
plot_B=ax1.plot(mm_scale,B, label='Field', color='b',lw=3)
ax2 = ax1.twinx()
plot_ne=ax2.plot(mm_scale,avg_ne_scaled/1e17, label='Line integrated density', color='g', lw=3)

ax1.set_xlim([-5,5])
ax1.set_xticks(np.arange(-5,5+1,1))
ax1.set_xlabel('Distance from centre (mm)', fontsize=40)
ax1.set_ylim([-10,10])
ax1.set_ylabel('Average Magnetic Field (T)',fontsize=36, color='b')
ax2.set_ylim([0,14])
ax2.set_ylabel(r'Line integrated electron density ($\times10^{17}\,cm^{-2}$)',fontsize=36, color='g')
ax1.tick_params(labelsize=34, pad=5, length=10, width=2)
ax2.tick_params(labelsize=34, pad=5, length=10, width=2)
#ax1.legend(loc=2, fontsize=26)
#ax2.legend(loc=0, fontsize=26)
ax1.set_title('s0513_15: Magnetic field and Electron density',fontsize=36)

```

```python
import csv

filename='s0513_15 line profiles'
export=zip(mm_scale, avg_ne_scaled, avg_rot_clipped, B)
headings=('Distance','n_e*L','rotation angle', 'B')
units=('mm','cm**-2', 'degrees', 'Tesla')
with open(filename+'.dat', 'wb') as f:
    writer = csv.writer(f)
    writer.writerow(headings)
    writer.writerow(units)
    for f in export:
        writer.writerow(f)
```
